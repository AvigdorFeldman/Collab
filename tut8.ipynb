```python
import nltk
from nltk.chat.util import Chat, reflections

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')

# Define some patterns and responses
patterns = [
    (r'hi|hello|hey', ['Hello!', 'Hi there!', 'Hey!']),
    (r'how are you\??', ['I\'m good, thank you!', 'I\'m doing well, thanks for asking!']),
    (r'what is your name\??', ['You can call me ChatGPT.', 'I go by the name ChatGPT.']),
    (r'(.*) your name\??', ['You can call me ChatGPT.', 'I go by the name ChatGPT.']),
    (r'bye|goodbye', ['Goodbye!', 'See you later!', 'Take care!']),
    (r'what can you do\??', ['I can chat with you and answer simple questions!', 'I\'m a simple chatbot who can engage in basic conversation.']),
    (r'thank you|thanks', ['You\'re welcome!', 'No problem!', 'My pleasure!']),
    (r'how old are you\??', ['I\'m a computer program, so I don\'t have an age!', 'Age is just a number for chatbots!']),
    (r'(.*) weather (.*)', ['I cannot check the weather, but I hope it\'s nice where you are!']),
    (r'What are the most common plant diseases\??', ['The most common plant diseases include powdery mildew, which causes a white coating on leaves, leaf spot, which creates dark spots on foliage, blight, leading to rapid browning and plant death, root rot caused by excess moisture, and rust, which appears as orange or brown patches on leaves.']),
    (r'What are the care methods for a typical plant disease\??(.*)', ['Care methods for typical plant diseases include removing infected plant parts, improving air circulation and sunlight, avoiding overwatering, using disease-resistant plant varieties, and applying appropriate fungicides or treatments when necessary.']),
    (r'How can overwatering harm plants?', ['Overwatering can suffocate roots by reducing oxygen in the soil, leading to root rot, yellowing leaves, slowed growth, and increased risk of fungal diseases.']),
    (r'Why is location important for plant health?', ['A plantâ€™s location affects sunlight, temperature, airflow, and soil drainage, all of which influence growth, disease resistance, and overall health.']),
]

# Create a chatbot
chatbot = Chat(patterns, reflections)

# Start the conversation
def start_chat():
    print("Hello! I'm ChatGPT. How can I help you today? (Type 'quit' to exit)")

    while True:
        user_input = input("You: ").lower()
        if user_input == 'quit':
            print("ChatGPT: Goodbye! Have a great day!")
            break

        response = chatbot.respond(user_input)
        print("ChatGPT:", response)

if __name__ == "__main__":
    start_chat()
```
