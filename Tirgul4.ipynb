```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'

import requests
import pandas as pd
from ipywidgets import widgets, VBox, Output
from IPython.display import display
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)

response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

data_df = pd.DataFrame(data['result']['records'])
data_df.head()
output_area = Output()

def get_years(df):
  moed_aliya_lakvish_all = df.get('moed_aliya_lakvish')
  years = set()
  for moed_aliya_lakvish in moed_aliya_lakvish_all:
    if(moed_aliya_lakvish is not None):
      years.add(moed_aliya_lakvish.split('-')[0])
  return sorted(list(years))

def get_unique_car_manufacturers(df, years):
  nums = []
  for year in years:
    df_filtered = df[df['moed_aliya_lakvish'].notna()].copy()
    df_filtered = df_filtered[df_filtered['moed_aliya_lakvish'].str.contains(year)]
    unique_car_manufacturers = df_filtered['tozeret_nm'].nunique()
    nums.append(unique_car_manufacturers)
  return nums

def yearly_bar_chart(df):

  # Extract year and convert to integer
  years = get_years(df)
  numOfCars = get_unique_car_manufacturers(df, years)
  plt.figure(figsize=(12, 6))
  plt.bar(years, numOfCars)
  plt.xlabel('Year')
  plt.ylabel('Number of Unique Car Manufacturers')
  plt.title('Number of Unique Car Manufacturers by Year')
  plt.show()


tab2_content = widgets.Output()
with tab2_content:
  print("Data Table:")
  display(data_df)  # Summary statistics

tab1_content = widgets.Output()
with tab1_content:
    print("Statisticks Overview:")
    display(data_df.describe())  # Summary statistics
tab3_content = widgets.Output()
with tab3_content:
    print("Year Bar Chart:")
    yearly_bar_chart(data_df)


# Create Tabs
tabs = widgets.Tab(children=[tab1_content, tab2_content,tab3_content])
tabs.set_title(0, 'Tab 1: Raw Data')
tabs.set_title(1, 'Tab 2: Tabled Data')
tabs.set_title(2, 'Tab 3: Year Bar Chart')




# Display Tabs
display(tabs)
```
