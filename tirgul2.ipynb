```python
# @title Students
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
from google.colab import drive
import json # Ensure json module is imported
drive.mount('/content/drive')

dropdown = widgets.Dropdown(
  options = ['', 'Shahar', 'Avigdor','Niran']
)
display(dropdown)

def get_information(name):
    # Open and load the JSON file
    with open('/content/drive/My Drive/students.json', 'r') as fid:
        students = json.load(fid)
    for student in students:
        if student['firstName'] == name:
            return student['firstName'], student['lastName'], student['Email'] , student['Courses'], student['Link']
    return None


def handle_change(change):
  clear_output(wait=True)
  display(dropdown)
  student_info = get_information(change.new)
  if student_info:
    html_content = f"""
      <html>
        <head>
          <style>
            th {{
              text-align: left;
            }}
            td {{
              text-align: left;
            }}
          </style>
        </head>
        <body>
          <table>
            <tr>
              <th>First Name:</th>
              <td>{student_info[0]}</td>
            </tr>
            <tr>
              <th>Family Name:</th>
              <td>{student_info[1]}</td>
            </tr>
            <tr>
              <th>Email:</th>
              <td>{student_info[2]}</td>
            </tr>
            <tr>
              <th>Courses:</th>
              <td>{', '.join(student_info[3])}</td>
            </tr>
            <tr>
              <th>Link:</th>
              <td><a href="{student_info[4]}" target="_blank">{student_info[4]}</a></td>
            </tr>
          </table>
        </body>
      </html>
        """
    display(HTML(html_content))

dropdown.observe(handle_change, names='value')
```
