```python
# @title Students
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
from google.colab import drive
import json
drive.mount('/content/drive')

dropdown = widgets.Dropdown(
  options = ['', 'Shahar', 'Avigdor','Niran']
)
display(dropdown)
favorite_show_input = widgets.Textarea(placeholder='Enter your favorite show')
update_button = widgets.Button(description='Update') # Re-adding update button definition

def get_students_json():
  with open('/content/drive/My Drive/students.json', 'r') as fid:
      students = json.load(fid)
  return students

def get_information(name):
    students = get_students_json()
    for student in students:
        if student['firstName'] == name:
            favorite_show = student.get('FavoriteShow', '')
            return student['firstName'], student['lastName'], student['Email'] , student['Courses'], student['Link'], favorite_show
    return None

def update_student_json(students):
    with open('/content/drive/My Drive/students.json', 'w') as fid:
        json.dump(students, fid)

def update_student_info(button_event):
    selected_name = dropdown.value
    students = get_students_json()
    for student in students:
        if student['firstName'] == selected_name:
          student['FavoriteShow'] = favorite_show_input.value
          break
    update_student_json(students)
    handle_change({'new': selected_name})

def handle_change(change):
  clear_output(wait=True)
  display(dropdown)

  student_info = get_information(change['new'])
  if student_info:
    html_content = f"""
      <html>
        <head>
          <style>
            th {{
              text-align: left;
            }}
            td {{
              text-align: left;
            }}
          </style>
        </head>
        <body>
          <table>
            <tr>
              <th>First Name:</th>
              <td>{student_info[0]}</td>
            </tr>
            <tr>
              <th>Family Name:</th>
              <td>{student_info[1]}</td>
            </tr>
            <tr>
              <th>Email:</th>
              <td>{student_info[2]}</td>
            </tr>
            <tr>
              <th>Courses:</th>
              <td>{', '.join(student_info[3])}</td>
            </tr>
            <tr>
              <th>Link:</th>
              <td><a href="{student_info[4]}" target="_blank">{student_info[4]}</a></td>
            </tr>
          </table>
        </body>
      </html>
        """
    display(HTML(html_content))
    favorite_show_input.value = student_info[5] if len(student_info) >= 5 and student_info[5] else ''
    display(favorite_show_input)
    display(update_button)
dropdown.observe(handle_change, names='value')
update_button.on_click(update_student_info)
```
