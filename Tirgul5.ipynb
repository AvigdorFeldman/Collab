```python
from firebase import firebase
import os
# חיבור לפיירבייס
FBconn = firebase.FirebaseApplication(
    'https://tirgul5-7c2cf-default-rtdb.firebaseio.com/',
    None
)

def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

def get_word_count(word):
    return FBconn.get('/', word)

def set_word_count(word, count):
    FBconn.put('/', word, count)

def delete_word(word):
    FBconn.delete('/', word)

def get_all_words():
    return FBconn.get('/', None)

while True:
    clear_screen()
    print("\n--- תפריט ---")
    print("1. הוספת מילה בודדת")
    print("2. הוספת טקסט שלם לניתוח")
    print("3. עדכון מספר ההופעות של מילה")
    print("4. מחיקת מילה")
    print("5. צפייה בכל המילים השמורות")
    print("6. יציאה")

    choice = input("בחר אופציה: ")

    # 1. הוספת מילה בודדת
    if choice == "1":
        word = input("הכנס מילה: ").strip()
        count = get_word_count(word)

        if count is None:
            set_word_count(word, 1)
            print(f"המילה '{word}' נוספה עם כמות 1.")
        else:
            set_word_count(word, count + 1)
            print(f"המילה '{word}' עודכנה ל-{count + 1}.")

    # 2. הוספת טקסט שלם
    elif choice == "2":
        text = input("הכנס טקסט: ")

        for word in text.split():
            word = word.strip()
            if word:
                count = get_word_count(word)
                if count is None:
                    set_word_count(word, 1)
                else:
                    set_word_count(word, count + 1)

        print("הטקסט הוכנס ועובד בהצלחה.")

    # 3. עדכון מספר הופעות
    elif choice == "3":
        word = input("הכנס מילה לעדכון: ")
        count = get_word_count(word)

        if count is None:
            print("המילה לא קיימת!")
        else:
            new_val = int(input("הכנס ערך חדש: "))
            set_word_count(word, new_val)
            print(f"המילה '{word}' עודכנה ל-{new_val}.")

    # 4. מחיקת מילה
    elif choice == "4":
        word = input("הכנס מילה למחיקה: ")
        count = get_word_count(word)

        if count is None:
            print("המילה לא קיימת!")
        else:
            delete_word(word)
            print(f"המילה '{word}' נמחקה מהמאגר.")

    # 5. הצגת כל המילים
    elif choice == "5":
        data = get_all_words()
        print("\n--- המילים במערכת ---")
        if not data:
            print("אין מילים שמורות.")
        else:
            for w, c in data.items():
                print(f"{w} : {c}")

    # 6. יציאה
    elif choice == "6":
        print("להתראות!")
        break

    else:
        print("בחירה לא תקינה.")
```
